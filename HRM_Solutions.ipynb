11. Feature Engineering

11.1 Label Encoding: Converting categorical features into numerical

```python
from sklearn.preprocessing import LabelEncoder
label_encoder= LabelEncoder()
```

```python
Emp_data1['salary']=label_encoder.fit_transform(Emp_data1['salary'])
Emp_data1['Department']=label_encoder.fit_transform(Emp_data1['Department'])
Emp_data1.head()
```

11.2 Dependent (Target) and Independent Variable

```python
x= Emp_data1.drop('left', axis=1)
y= Emp_data1['left']
```

11.3 Splitting data into Training and Testing

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test= train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
x_train.shape
```

11.4 Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
std_scaler= StandardScaler()
```

```python
xtrain_scaled= std_scaler.fit_transform(x_train)
xtest_scaled= std_scaler.transform(x_test)
```

```python
xtrain_scaled
xtest_scaled
```

12. Model Development

12.1 Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
Random_forest_model= RandomForestClassifier()
```

12.2 Model Training

```python
Random_forest_model.fit(xtrain_scaled, y_train)
```

12.3 Model Prediction

```python
y_pred= Random_forest_model.predict(xtest_scaled)
y_pred
```
